For a cluster with arbitrary shape parameter $f$, local F-actin production (in a n-cluster system) near the cluster is given by 

$$F_{local} = C - \beta K(f)^{1/2} n^{1/6} $$


where $$ \beta = \sqrt{\frac{1}{\alpha} (\frac{k2}{k1})} $$

```python
import numpy as np
from numpy import array, exp, sqrt, log  
import matplotlib.pyplot as plt


font = {'family' : 'Arial',
        'size'   : 16}

plt.rc('font', **font)

def calc_oblate_factor(hS):
    if hS == 1:
        Kf = 1
        return Kf 
    else:
        f = hS**(2/3)
        e = sqrt((1 - f**3))
        p = (1+e)/(1-e)
        Kf = 0.5*(1/f + ((f**2)/(2*e)) * log(p))
        return Kf

n = np.arange(1,1001,1)

# Initial conditions
G0 = 100
F0 = 0.01 
C = G0 + F0

# rate parameters
k1 = 0.001
k2 = 1
a = 1

title = f'C = {C}, k1 = {k1}, k2 = {k2}, a = {a}'

fn_Kf = lambda n, a, C, k1, k2, hS: C - sqrt(k2/(k1*a))* (calc_oblate_factor(hS))**0.5 * n**(1/6)

hSList = [1, 0.8, 0.6, 0.4]
colors = ['r', 'b', 'g', 'm']
ms = 5
for hS, color in zip(hSList, colors):
    Fn_f = fn_Kf(n, a=a, C=C, k1=k1, k2=k2, hS=hS)
    Fn_f[Fn_f < 0] = 0  # to avoid numerical unstable solutions
    plt.plot(np.log10(n), np.ma.log10(Fn_f), f'{color}o--', ms=ms, label=f'hS = {hS}')
    plt.legend()

plt.xticks([0,1,2,3], labels=['10$^0$','10$^1$','10$^2$','10$^3$'], fontsize=16)
plt.yticks([-2,-1,0,1,2], labels=['10$^{-2}$','10$^{-1}$','10$^0$','10$^1$','10$^2$'], fontsize=16 )    
plt.xlabel('Cluster number (n)', fontweight='bold')
plt.ylabel('Local F (F$_n$)')
plt.show()

for hS, color in zip(hSList, colors):
    Fn_f = fn_Kf(n, a=a, C=C, k1=k1, k2=k2, hS=hS)
    Fn_f[Fn_f < 0] = 0  # to avoid numerical unstable solutions
    plt.plot(np.log10(n), np.ma.log10(n*Fn_f), f'{color}o--', ms=ms, label=f'hS = {hS}')
    plt.legend()

plt.xticks([0,1,2,3], labels=['10$^0$','10$^1$','10$^2$','10$^3$'], fontsize=16)

plt.yticks([0,1,2,3,4], labels=['10$^0$','10$^1$','10$^2$', '10$^3$','10$^4$',], fontsize=16) 

plt.xlabel('Cluster number (n)', fontweight='bold')
plt.ylabel('Global F (F$_n$)')
plt.show()
```
